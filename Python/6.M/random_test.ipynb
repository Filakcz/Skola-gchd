```python
import random
import matplotlib.pyplot as plt

pocet = 1000000  
orel = 0

historie = []

for i in range(1, pocet):
    if random.choice([True, False]):
        orel += 1
    if i % 1000 == 0:
        historie.append(100*orel / i)


plt.plot(range(1, pocet//1000), historie)
plt.axhline(50, color="red", linestyle="--", label="50 %")

plt.xlabel("Počet hodů v tisích")
plt.ylabel("Podíl orla vůči panně")
plt.title("Podíl orla vůči panně při 1 milionu hodů")
plt.legend()
plt.show()
```

```python
vysledky = {}
for i in range(1, 7):
    vysledky[i] = 0
        
for i in range(pocet):
    hod = random.randint(1, 6)
    vysledky[hod] += 1

plt.bar(vysledky.keys(), vysledky.values())
plt.axhline(pocet/6, color="red", linestyle="--", label="Očekávaný počet")

plt.yscale("log")

plt.xlabel("Hodnota")
plt.ylabel("Počet výskytů (logaritmické měřítko)")
plt.title("Rozdělení hodů kostky s 6 stěny při 1 milionu hodů")
plt.legend()

plt.show()
```
