```python
from matplotlib import pyplot as plt
import pandas as pd

df8M = pd.read_csv("doppler_8M.csv")
df8J = pd.read_csv("doppler_8J.csv")
# Create a figure and arrange subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 5), sharey=True)  # 1 row, 3 columns

# Plot the histograms
axs[0][0].hist(df8M["CJL_body"], bins=10, color='blue', alpha=0.7)
axs[0][0].set_title("esk jazyk a literatura")
axs[0][0].set_xlabel("Poet zskanch bod")
axs[0][0].set_ylabel("Poet uchaze s danm skre")

axs[0][1].hist(df8M["MA_body"], bins=10, color='green', alpha=0.7)
axs[0][1].set_title("matematika a jej aplikace")
axs[0][1].set_xlabel("Poet zskanch bod")

# Plot the histograms
axs[1][0].hist(df8J["CJL_body"], bins=10, color='blue', alpha=0.7)
axs[1][0].set_title("esk jazyk a literatura")
axs[1][0].set_xlabel("Poet zskanch bod")
axs[1][0].set_ylabel("Poet uchaze s danm skre")

axs[1][1].hist(df8J["MA_body"], bins=10, color='green', alpha=0.7)
axs[1][1].set_title("matematika a jej aplikace")
axs[1][1].set_xlabel("Poet zskanch bod")

def show_measure_of_location(data_column, plot):
    min_value = data_column.min()
    max_value = data_column.max()
    mean_value = data_column.mean()
    median_value = data_column.median()
    mode_value = data_column.mode()[0]

    plot.axvline(x=min_value, color="gray")
    plot.axvline(x=max_value, color="gray")


    plot.axvline(x=mean_value, color="yellow")
    plot.axvline(x=median_value, color="green")

    
    plot.axvline(x=mode_value, color="red") # nejastj hodnota


show_measure_of_location(df8M["CJL_body"],axs[0][0])
show_measure_of_location(df8M["MA_body"],axs[0][1])
show_measure_of_location(df8J["CJL_body"],axs[1][0])
show_measure_of_location(df8J["MA_body"],axs[1][1])


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()


```

```python
def show_measure_of_variability(obor, data_column, nazev_columnu):
    range = data_column.max() - data_column.min()
    mean = data_column.mean()
    standard_deviation = data_column.std()
    variation = standard_deviation**2

    print(f"{obor} - {nazev_columnu}: \n\trange: {range}\n\tmean: {mean}\n\tstandard deviation: {standard_deviation}\n\tvariation: {variation}")
    print()

show_measure_of_variability("8.M", df8M["CJL_body"], "CJL_body")
show_measure_of_variability("8.M", df8M["MA_body"], "MA_body")
show_measure_of_variability("8.J", df8J["CJL_body"], "CJL_body")
show_measure_of_variability("8.J", df8J["MA_body"], "MA_body")
```
